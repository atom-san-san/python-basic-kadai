```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name, "は大人です")
        else:
            print(self.name, "は大人ではありません")


# Humanクラスのインスタンスを複数生成
humans = [
    Human("Taro", 25),
    Human("Hanako", 18),
    Human("Jiro", 20)
]

# リストの要素数分だけメソッドを呼び出す
for human in humans:
    human.check_adult()
```
