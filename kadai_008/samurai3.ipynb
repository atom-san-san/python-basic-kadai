```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 正の整数（1以上の非常に大きな数）を生成
num = random.randint(1, 10**9)

# 変数numの値を出力
print(num)

# 判定結果を変数に格納
if num % 3 == 0 and num % 5 == 0:
    print("FizzBuzz")
elif num % 3 == 0:
    print("Fizz")
elif num % 5 == 0:
    print("Buzz")
else:
    print(str(num))
```
