```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 正の整数（1以上の非常に大きな数）を生成
var = random.randint(1, 10**9)

# 変数varの値を出力
print(var)

# 判定結果を変数に格納
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
elif var % 3 == 0:
    print("Fizz")
elif var % 5 == 0:
    print("Buzz")
else:
    print(str(var))
```
